# IMPORT LIBRARY

```python
import numpy
import pandas
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils import np_utils
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
```

```python
# Maintain model output validity
seed = 7
numpy.random.seed(seed)
```

```python
dataset = pandas.read_csv("DATA/iris.csv")
dataset.head()
```

```python
X = dataset.values[:,0:4].astype(float)
Y = dataset.values[:,4]

print(f"shape of X: {X.shape}")
print(f"shape of Y: {Y.shape}")
```

```python
# Convert target variable values to quantities
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
dummy_y = np_utils.to_categorical(encoded_Y)
```

# CREATE MLP MODEL

```python
def baseline_model():

    model = Sequential()
    model.add(Dense(6, input_dim=4, kernel_initializer= 'normal' , activation= 'relu'))
    model.add(Dense(4, kernel_initializer= 'normal' , activation= 'relu' ))
    model.add(Dense(3, kernel_initializer= 'normal' , activation= 'softmax'))

    model.compile(loss= 'categorical_crossentropy' , optimizer= 'adam' , metrics=['accuracy'])
    return model
```

```python
estimator = KerasClassifier(build_fn=baseline_model, epochs=200, batch_size=32)
```

```python
kfold = KFold(n_splits=10, shuffle=True, random_state=seed)
results = cross_val_score(estimator, X, dummy_y, cv=kfold)
```

# MODEL ACCURACY

```python
print("Accuracy: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```
